**>> DATA CLEANING**

1st column : order_id

```python
# Check for duplicated in column "order_id"
duplicate = df["order_id"].duplicated().sum()
print(f"The number of duplicated values is : {duplicate}")
```

```python
# Inspect duplicated order_id to determine if rows are identical or contain inconsistencies
df[df["order_id"].duplicated(keep=False)]
```

```python
# Remove duplicate rows based on order_id
df = df.drop_duplicates(subset="order_id")

# Check again for duplicated values
duplicate = df["order_id"].duplicated().sum()
print(f"The number of duplicated values is : {duplicate}")
```

2nd column : customer_name

```python
# Check missing values in customer_name
missing_values = df["customer_name"].isnull().sum()
print(f"The number of missing values is : {missing_values}")
```

```python
print(df)

df["customer_name"].value_counts()

# Standardize customer_name formatting
df["customer_name"] = df["customer_name"].str.strip().str.title()

print(df)
```

```python
# SReplace missing names (NaN) with placeholder ("Unknown")
df["customer_name"] = df["customer_name"].fillna("Unknown")
print(df)

df = df[df["customer_name"] != "Unknown"]
print(df)
```

3rd column : email

```python
# Check frequency emails
df["email"].value_counts()
```

```python
# Check for NOT valid emails (missing '@' or '.')
invalid_emails = df[
    ~df["email"].str.contains("@") |
    ~df["email"].str.contains(r"\.")
]

print(invalid_emails)
```

```python
# Set invalid email formats to NaN to avoid introducing incorrect information
df.loc[
    ~(
        df["email"].str.contains("@") &
        df["email"].str.contains(r"\.")
    ),
    "email"
] = np.nan

print(df["email"])
```

```python
df["email"] = df["email"].str.lower()
df
```

4th column : country

```python
# Check unique values in country column
df["country"].value_counts()
```

```python
# Normalize text (lowercase + remove extra spaces)
df["country"] = df["country"].str.lower().str.strip()

df["country"]
```

```python
# Map all variations to 'italy'

df["country"] = df["country"].replace({
    "italia": "italy",
    "it" : "italy",
    "italya" : "italy",
}
)

df["country"]
```

```python
# Standardize final format (capitalize)
df["country"] = df["country"].str.title()

df["country"].value_counts()
```

5th column : order_date

```python
# Check raw values of order_date
df["order_date"]
```

```python
# Normalize date separators
df["order_date"] = df["order_date"].str.replace("/", "-", regex=False)
df["order_date"] = df["order_date"].str.replace(".", "-", regex=False)
df["order_date"]
```

```python
# Due to inconsistent date formats, a new cleaned column "order_date_clean" is created with manually standardized values.

df["order_date_clean"] = [
    "12-01-2025",
    "13-01-2025",
    "14-01-2025",
    "14-01-2025",
    "15-01-2025",
    "15-01-2025",
    "16-01-2025",
    None
]

df["order_date_clean"]
```

```python
df["order_date_clean"] = pd.to_datetime(
    df["order_date_clean"],
    dayfirst=True
)


df["order_date_clean"]
df
```

```python
# Replace original column
df["order_date"] = df["order_date_clean"]

# Drop helper column
df = df.drop(columns=["order_date_clean"])
```

6th column : revenue

```python
df["revenue"]
```

```python
df["revenue"] = df["revenue"].str.replace("€", "", regex=False)
df["revenue"]
```

```python
# Convert to numeric (invalid values → NaN)
df["revenue"] = pd.to_numeric(df["revenue"], errors="coerce")
print(df["revenue"])

media = df["revenue"].mean()
print(media)
```

```python
#Rename column "revenue" with "revenue_Euro(€)"
df.rename(columns={"revenue": "revenue_Euro(€)"}, inplace=True)
df
```

7ty column : age

```python
df["age"]
```

```python
df.loc[(df["age"] < 0) | (df["age"] > 100), "age"] = None
df["age"]

df["age"] = df["age"].astype("Int64")
df["age"]
```

**>> FINAL RESULT**

```python
df.reset_index(drop=True, inplace=True)
df
```

# **# Final cleaned dataset ready for analysis**

```python
# Exporting the 'cleaned' version of the dataset
df.to_csv("cleaned_ecommerce_dataset.csv", index=False)
```
